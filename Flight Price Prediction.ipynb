```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('flights.csv')
df
```

```python
airline = df['airliner']
airline = pd.get_dummies(airline)
airline.head()
```

```python
list1 = ['origin', 'destination']
for l in list1:
    print(df[[l]].value_counts(),'\n')
```

```python
df = pd.get_dummies(data = df, columns = list1)
df
```

```python
df = pd.concat([df, airline], axis = 1)
df.drop(['airliner', 'airport_origin_name', 'airport_destination_name'], axis = 1, inplace = True)
df
```

```python
df['day'] = pd.to_datetime(df['date']).dt.day
df['month'] = pd.to_datetime(df['date']).dt.month
df.drop(['date'], axis = 1, inplace = True)
```

```python
df['depart_hr'] = pd.to_datetime(df['depart_time'].astype(str)).dt.hour
df['depart_min'] = pd.to_datetime(df['depart_time'].astype(str)).dt.minute
df['arr_hr'] = pd.to_datetime(df['arrive_time'].astype(str)).dt.hour
df['arr_min'] = pd.to_datetime(df['arrive_time'].astype(str)).dt.minute
df.drop(['depart_time', 'arrive_time', 'duration_min'], axis = 1, inplace = True)
```

# <h2>Feature Selection/Importance</h2>

```python
x = df.drop(['price'], axis =1 )
y = df['price']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data = sc.fit_transform(x)
data[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 123)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# <h2><center>ExtraTreesRegressor Model</center></h2>

```python
from sklearn.ensemble import ExtraTreesRegressor

model = ExtraTreesRegressor()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

def metrics(y_true, y_pred):
    print(f'RMSE:', mean_squared_error(y_test, y_pred)**0.5)
    print(f'R2 score:', r2_score(y_test, y_pred))
def accuracy(y_true, y_pred):
    errors = abs(y_pred - y_true)
    mape = 100 * np.mean(errors / y_true)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
```

```python
y_pred = model.predict(X_test)
```

```python
metrics(y_test, y_pred)
```

```python
accuracy(y_test, y_pred)
```

# <h2><center>Linear Regression Model</center></h2>

```python
from sklearn.linear_model import LinearRegression

linearmodel = LinearRegression()
linearmodel.fit(X_train, y_train)
```

```python
y_linear_pred = linearmodel.predict(X_test)
```

```python
metrics(y_test, y_linear_pred)
```

```python
accuracy(y_test, y_linear_pred)
```

```python
sns.distplot(y_test-y_linear_pred)
plt.show()
```

# <h2><center>Random Forest Model</center></h2>

```python
from sklearn.ensemble import RandomForestRegressor
randomforest_model = RandomForestRegressor(n_estimators=100, min_samples_split = 3)
randomforest_model.fit(X_train, y_train)
```

```python
pred_rf = randomforest_model.predict(X_test)
```

```python
metrics(y_test, pred_rf)
```

```python
accuracy(y_test, pred_rf)
```
